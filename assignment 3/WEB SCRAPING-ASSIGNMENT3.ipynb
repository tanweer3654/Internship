```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#lets first connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")
#opening  page on automated chrome browser
driver.get("https://www.amazon.in/")
```

```python
# Getting User Input for Product to Search
product = input("Enter Product to Search: ")

#entering  in search field
search=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input" )
search.clear()
search.send_keys(product)
#clicking search
search=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search.click()



```

```python
product_url=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_url.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[70]/div/div/span/a[3]')
    time.sleep(2)
```

```python
len(product_url)
```

```python
brand_list=[]
product_list=[]
price_list=[]
return_list=[]
delivery_list=[]
avilability_list=[]

for url in product_url:
    driver.get(url)
    time.sleep(2)
    try:
        brand=driver.find_element(By.XPATH,'//div[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        brand_list.append(brand.text)
    except NoSuchElementException:
        brand_list.append('-')
    try:
        product=driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        product_list.append(product.text)
    except NoSuchElementException:
        product_list.append('-')
    try:
        price=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price_list.append(price.text)
    except NoSuchElementException:
        price_list.append('-')
    try:
        returnt=driver.find_element(By.XPATH,'//div[@id="icon-farm-container"]/div/div[3]/span/div[2]/a')
        return_list.append(returnt.text)
    except NoSuchElementException:
        return_list.append('-')
    try:
        delivery=driver.find_element(By.XPATH,'//div[@id="deliveryBlockContainer"]/div/div/div/div/span/span')
        delivery_list.append(delivery.text)
    except NoSuchElementException:
        delivery_list.append('-')
    try:
        availability=driver.find_element(By.XPATH,'//div[@class="a-section"]/div/span')
        avilability_list.append(availability.text)
    except NoSuchElementException:
        avilability_list.append('-')
```

```python
print(len(brand_list),len(product_list),len(price_list),len(return_list),len(delivery_list),len(avilability_list))
```

```python
#making dataframe
df=pd.DataFrame({"BRAND":brand_list,"PRODUCT DESCRIPTION":product_list,"PRICE":price_list,"RETURN/EXCHANGE":return_list,"DELIVERY":delivery_list,"AVAILABILITY":avilability_list})
df
```

```python
# saving the dataframe
df.to_csv('amazonproduct.csv')
```

```python
#scrapping the data
rank_list=[]
name_list=[]
networth_list=[]
age_list=[]
source_list=[]
citizenship_list=[]

#scrapping job title from webpage
rank_tags=driver.find_elements(By.XPATH,'//div[@class="scrolly-table"]/table/tbody/tr/td[2]')
for i in rank_tags:
    rank=i.text
    rank_list.append(rank)
name_tags=driver.find_elements(By.XPATH,'//div[@class="scrolly-table"]/table/tbody/tr/td[3]')
for i in name_tags:
    name=i.text
    name_list.append(name)
nw_tags=driver.find_elements(By.XPATH,'//div[@class="scrolly-table"]/table/tbody/tr/td[4]')
for i in nw_tags:
    networth=i.text
    networth_list.append(networth)
age_tags=driver.find_elements(By.XPATH,'//div[@class="scrolly-table"]/table/tbody/tr/td[6]')
for i in age_tags:
    age=i.text
    age_list.append(age)
src_tags=driver.find_elements(By.XPATH,'//div[@class="scrolly-table"]/table/tbody/tr/td[7]')
for i in src_tags:
    src=i.text
    source_list.append(src)
ctz_tags=driver.find_elements(By.XPATH,'//div[@class="scrolly-table"]/table/tbody/tr/td[8]')
for i in ctz_tags:
    ctz=i.text
    citizenship_list.append(ctz)
    
    
```
