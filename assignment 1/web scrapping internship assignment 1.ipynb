```python

# sending a GET request to the IMDB website
response = requests.get('https://www.imdb.com/chart/top')

# parsing the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# initializing a list to store the data
movies = []

# looping through all the rows in the table
for row in soup.find_all('td', {'class': 'titleColumn'}):
    # extracting the name, rating, and year of release
    name = row.find('a').text
    rating = row.find_next_sibling().find('strong').text
    year = row.find('span').text[1:5]
    
    # adding the data to the list
    movies.append((name, rating, year))

# creating the data frame
df = pd.DataFrame(movies, columns=['Name', 'Rating', 'Year'])

# printing the data frame
print(df)
```

```python


# URL for ODI rankings
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"

# Send a request to the URL and get the HTML content
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

# Find the table with the rankings information
table = soup.find("table", class_="table")

# Get the rows of the table
rows = table.find_all("tr")

# Create lists to store the rankings information
teams = []
matches = []
points = []
ratings = []

# Loop through the rows and extract the rankings information
for row in rows[1:]:
    cells = row.find_all("td")
    teams.append(cells[1].text.strip())
    matches.append(cells[2].text.strip())
    points.append(cells[3].text.strip())
    ratings.append(cells[4].text.strip())

# Create a data frame with the rankings information
df = pd.DataFrame({"Team": teams, "Matches": matches, "Points": points, "Rating": ratings})

# Filter the data frame to get the top 10 teams
df = df[:10]

# Display the data frame
print("Top 10 ODI Teams:")
print(df)

# Repeat the process for ODI batsmen and bowlers
# URL for ODI batsmen rankings
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"

# Send a request to the URL and get the HTML content
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

# Find the table with the rankings information
table = soup.find("table", class_="table")

# Get the rows of the table
rows = table.find_all("tr")

# Create lists to store the rankings information
names = []
teams = []
ratings = []

# Loop through the rows and extract the rankings information
for row in rows[1:]:
    cells = row.find_all("td")
    names.append(cells[1].text.strip())
    teams.append(cells[2].text.strip())
    ratings.append(cells[3].text.strip())

# Create a data frame with the rankings information
df = pd.DataFrame({"Name": names, "Team": teams, "Rating": ratings})

# Filter the data frame to get the top 10 batsmen
df = df[:10]

# Display the data frame
print("\nTop 10 ODI Batsmen:")
print(df)

# URL for ODI bowlers rankings
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"

# Send a request to the URL and get the HTML content
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

# Find the table with the rankings information
table = soup.find("table", class_="table")

# Get the rows of the table
rows = table.find_all("tr")

# Create lists to store the rankings information
names = []
teams = []
ratings = []

# Loop through the rows and extract the rankings information
for row in rows[1:]:
    cells = row.find_all("td")
    names.append(cells[1].text.strip())
    teams.append(cells[2].text.strip())
    ratings.append(cells[3].text.strip())

# Create a data frame with the rankings information
df = pd.DataFrame({"Name": names, "Team": teams, "Rating": ratings})

# Filter the data frame to get the top 10 bowlers
df = df[:10]

# Display the data frame
print("Top 10 ODI Bowlers:")
print(df)



```

```python

url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

headlines = []
times = []
links = []

for article in soup.find_all("div", class_="FeaturedCard-contentText"):
    headline = article.find("a").text
    headline = headline.strip()
    headlines.append(headline)

    time = article.find("span", class_="datetime")
    times.append(time)

    link = article.find("a")["href"]
    links.append(link)

df = pd.DataFrame({"Headline": headlines, "Time": times, "Link": links})
print(df)
```

```python
# URL to scrape
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")

# Find all the articles on the page
articles = soup.find_all("li", class_="sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs")

# Create a list to store the details of the articles
articles_list = []

# Loop through each article
for article in articles:
    # Extract the title, authors, published date, and URL of the article
    title = article.find("a").text
    authors = article.find("span", class_="sc-1w3fpd7-0 dnCnAO").text
    published_date = article.find("span",class_="sc-1thf9ly-2 dvggWt")
    article_url = article.find("a")["href"]
    
    # Add the details to the list
    articles_list.append({
        "Title": title,
        "Authors": authors,
        "Published Date": published_date,
        "URL": article_url
    })

# Convert the list to a Pandas DataFrame
articles_df = pd.DataFrame(articles_list)

# Print the DataFrame
print(articles_df)
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
rname=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    rname.append(i.text)
cname=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cname.append(i.text.split('|')[1])
loc=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):    
    rating.append(i.text)
rate=rating[0:9]
images=[]
for i in soup.find_all('img',class_="no-img"):
    images.append(i.get("data-src"))

#making dataframe

df=pd.DataFrame({"Restaurant Name":rname,"Cuisine":cname,"Location":loc,"Rating":rate,"Images_url":images})
df
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
pname=[]
for i in soup.find_all('div',class_="presidentListing"):
    pname.append(i.text)
pname
```

```python

formatted_presidents = []

for president in pname:
    name = president.split("\n")[1].split("(")[0].strip()
    term = president.split("Term of Office:")[1].split("\n")[0].strip()
    formatted_presidents.append((name, term))

df = pd.DataFrame(formatted_presidents, columns=['Name', 'Term of Office'])
print(df)
```
