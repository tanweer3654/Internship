```python
#entering designation and location
designation=driver.find_element(By.CLASS_NAME,"suggestor-input" )
designation.send_keys("Data Analyst")

location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys("Bangalore")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[6]")
search.click()
```

```python
#scrapping the data
job_title=[]
job_location=[]
company_name=[]
experience_required=[]

#scrapping job title from webpage
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
#scrapping job location
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
#scrapping comapny name
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
#scapping experience details
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
    
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
#Making dataframe

df=pd.DataFrame({"Title":job_title,"Location":job_location,"Company Name":company_name,"Experience":experience_required})
df
```

```python
Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.
Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to click on the List option from Dropdown menu on left side.
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.
```

```python
#lets first connect to the driver
driver=webdriver.Chrome(r"chromedriver.exe")
#opening  page on automated chrome browser
driver.get("https://www.motor1.com/")
```

```python
#clicking on drop down menu
dd=driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div/div/div[1]/div")
dd.click()
```

```python
#clicking on features
features=driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div[3]/ul/li[5]/button")
features.click()
```

```python
#clicking on list
l=driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div[3]/ul/li[6]/ul/li[1]/a")
l.click()
```

```python
#now clicking on 50 most expensive car
ec=driver.find_element(By.XPATH,"/html/body/div[3]/div[9]/div[1]/div[1]/div/div/div[2]/div/div[1]/h3/a")
ec.click()
```

```python
cn=[]
price=[]
#scrapping car name from webpage
cn_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in cn_tags:
    cname=i.text
    cn.append(cname)
#scrapping price
price_tags=driver.find_elements(By.XPATH,'//strong')
for i in price_tags:
    p=i.text
    price.append(p)
```

```python
dollar_amounts = []

for price in prices:
    parts = price.split(": ")
    dollar_amount = parts[1]
    dollar_amounts.append(dollar_amount)

print(dollar_amounts)
```

```python
#making datframe
df=pd.DataFrame({"CAR NAME":cn,"PRICE":dollar_amount})
df
```
